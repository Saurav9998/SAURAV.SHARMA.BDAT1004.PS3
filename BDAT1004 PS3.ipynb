# Question 1
Occupations
- Step 1. Import the necessary libraries
- Step 2. Import the dataset from this address.
- Step 3. Assign it to a variable called users
- Step 4. Discover what is the mean age per occupation
- Step 5. Discover the Male ratio per occupation and sort it from the most to the least
- Step 6. For each occupation, calculate the minimum and maximum ages
- Step 7. For each combination of occupation and sex, calculate the mean age
- Step 8. For each occupation present the percentage of women and men

```python
#step 1
import pandas as pd
import numpy as np
```

```python
#Step 2 & 3
users = pd.read_csv('https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user', sep = '|')
```

```python
users
```

```python
#Step 4
users.groupby('occupation')['age'].mean()
```

```python
#Step 5 - creating a count for number of males 
users['males_count'] = users.gender.apply(lambda x: True if x == 'M' else False)
```

```python
#step 5 - Discovering male ratio using sum of male count and total gender count
#         Also sorting it into order most to the least
(users.groupby('occupation').males_count.sum() / users.groupby('occupation').gender.count()).sort_values(ascending = False)
```

```python
# Step 6 
print('Minimum age per occupation\n')
print(users.groupby('occupation').age.min())
print('\nMaximum age per occupation\n')
print(users.groupby('occupation').age.max())
```

```python
#Step 7
users.groupby(['occupation' , 'gender']).age.mean()
```

```python
gender_count = users.groupby(['occupation' , 'gender']).gender.count()
gender_count
```

```python
#Step 8 
print('The percentage of women and men based on the occupation:')
totalcount_per_occupation = users.groupby('occupation').gender.count()
gender_count*100/totalcount_per_occupation
```

# Question 7
Create a vertical bar chart comparing the number of marriages and divorces per
capita in the U.S. between 1900, 1950, and 2000.
